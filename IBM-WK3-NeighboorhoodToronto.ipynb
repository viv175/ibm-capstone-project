#### Data about 'List of postal codes of Canada: M' is taken from https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M

```python
# Import the libraries required
import pandas as pd
```

```python
# Read data from the URL provided

postcodeM = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
postcodeM[0]
```

```python
# Setup the final result data frame in the right format
column_names = ['Postcode', 'Borough','Neighborhood']

df_postcodeM = pd.DataFrame(columns=column_names)
df_postcodeM
```

```python
# Read that data and parse it so its split in the right format and data is put into the right columns and rows

Count = 0
print("PostcodeM Shape is: ",postcodeM[0].shape)
#for row in postcodeM[0]:
#    for column in postcodeM[0]:
for row in range(20):
    for column in range(9):
        # pick data from a cell i.e. specific row & colulm
        data = postcodeM[0][column][row]
        # the first 3 letters are the post code
        postcode = data[0:3]   
        # after the first 3 letters we have the Borough and Neighborhood
        Neighborhood = data[3:]
        # Unless that text is 'Not assigned' perform the following action to process data
        if Neighborhood != 'Not assigned': 
            location = Neighborhood.rsplit("(")
            # Here we know that the Borough name is NOT 'Not assigned', check if the Neighborhood name is 'Not assigned'
            if location[1] == 'Not assigned':
                # if  Neighborhood name is 'Not assigned' assign it the Borough Name
                location[1] = location[0]
            # Append data into the data frame
            df_postcodeM = df_postcodeM.append({'Postcode': postcode,
                                    'Borough' : location[0],
                                    'Neighborhood': location[1].replace(")", " ").replace("/",",")
                                   }, ignore_index=True) 
        else: 
            # Section added for sanity check to ensure right data was going in data frame, commented once confirmed
            # print("Row", row,"Column:", column, "Data:", Neighborhood)
            # df_postcodeM = df_postcodeM.append({'Postcode': postcode,
            #                                    'Borough' : Neighborhood,
            #                                    'Neighborhood': Neighborhood
            #                                   }, ignore_index=True)
            Count += 1
    

print("Shape of the DataFrame is ",df_postcodeM.shape)
print("Not Assigned Count is", Count)
```

```python
# Check if there are dual entries of postcode and mderge the rows
# Note: Visual inspection has confirmed that this is not the case. I am not sure why the assignment description mentions it.
#       Its possible that this was the case earlier and has since been corrected

df_postcodeM = df_postcodeM.groupby('Postcode').agg({
                                                    'Borough' : 'first',
                                                    'Neighborhood':','.join}).reset_index()
```
